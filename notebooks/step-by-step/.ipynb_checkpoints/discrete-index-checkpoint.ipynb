```python
import numpy as np
import pymc3 as pm
import pandas as pd
import seaborn as sns

from aesara import tensor as at
from matplotlib import pyplot as plt

from scipy import stats as st

import aesara

%config InlineBackend.figure_format = "retina"
%matplotlib inline
```

```python
K = 19
x = np.random.normal(loc=2., scale=2., size=[50,])

def stick_breaking(betas):

    sticks = at.concatenate(
        [
            [1],
            at.cumprod(1 - betas[:-1]),
        ]
    )

    return betas * sticks


def stick_glueing(weights):
    
    denominator = at.cumsum(
        at.concatenate(
            [
                [1],
                -weights[:-1]
            ]
        )
    )
    
    return weights/denominator
```

```python
with pm.Model() as model:
    norm = pm.Normal("normal", 0, 1, shape=(K,), transform=pm.transforms.ordered, initval=np.arange(K) - K // K)
    
    betas = pm.Beta("vh", 1., 1., shape=(K,)) # not to order
    weights = pm.Deterministic("wh", stick_breaking(betas))
    
    discrete_obs = at.le(x, norm)
    index = pm.Categorical("index", weights, observed=discrete_obs)
    
    posterior = pm.sample(1000)
```

```python
_ = pm.plot_trace(posterior)
```
