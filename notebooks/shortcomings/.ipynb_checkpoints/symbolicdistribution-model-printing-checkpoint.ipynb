```python
import pymc as pm
import numpy as np

from aesara.tensor.random.op import RandomVariable

pm.__version__
```

```python
with pm.Model() as model:
    psi = pm.Gamma("psi", 1., 1.)
    mu = pm.Normal("mu", 0., 5.)
    x = pm.ZeroInflatedPoisson("x", psi, mu)
    y = pm.Deterministic("y", x + 1)
    
    normal_dist = pm.Normal.dist(mu=0.0, sigma=1.0)
    censored_normal = pm.Censored("censored_normal", normal_dist, lower=-1, upper=1)

pm.model_to_graphviz(model)
```

```python
isinstance(x.owner.op, pm.distributions.mixture.MarginalMixtureRV)
```

```python
censored_normal.owner
```

```python
with pm.Model() as norm_model:
    mu = pm.Normal("mu", 0., 5.)
    sigma = pm.Gamma("sigma", 1., 1.)
    norm = pm.Normal("norm", mu, sigma)
    
pm.model_to_graphviz(norm_model)
```

```python
isinstance(norm.owner.op, RandomVariable)
```

```python
with pm.Model() as mix_model:
    weights = pm.Dirichlet("weights", a=np.tile(1, 3))
    mu_1 = pm.Normal("mu_1", -2., 3.)
    mu_2 = pm.Normal("mu_2", 2., 3.)
    mix = pm.Mixture(
        "mix", 
        weights, 
        comp_dists=[
            pm.Normal.dist(mu_1, 3.), 
            pm.Normal.dist(mu_2, 3.),
        ],
    )
```
