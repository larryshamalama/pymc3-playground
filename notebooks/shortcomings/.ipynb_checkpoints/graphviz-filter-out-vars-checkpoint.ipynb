```python
import sys
sys.path.insert(0, "../../../pymc")

import pymc as pm
import numpy as np

from pymc.model_graph import ModelGraph
```

```python
with pm.Model() as mix_model:
    weights = pm.Dirichlet("weights", a=np.tile(1, 2))
    mu_1 = pm.Normal("mu_1", -2., 3.)
    mu_2 = pm.Normal("mu_2", 2., 3.)
    mix = pm.Mixture(
        "mix", 
        weights, 
        comp_dists=[
            pm.Normal.dist(mu_1, 3.), 
            pm.Normal.dist(mu_2, 3.),
        ],
    )
    
    a = pm.Normal("a")
    b = pm.Normal("b", mu=1 + mix)
    
pm.model_to_graphviz(mix_model, selected_vars=["mix"])
```
