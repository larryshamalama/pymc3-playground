```python
import pymc3

print(f"PyMC3 version: {pymc3.__version__}")
```

### With observations in `observed`

```python
with pymc3.Model() as model:
    rate = pymc3.Beta('rate', alpha=2, beta=2)
    input_var = pymc3.Data('input_var', [1, 2, 3])
    output_var = pymc3.Binomial('output_var', n=input_var, p=rate, observed=[1, 1, 1])
pymc3.model_to_graphviz(model)
```

### Without observations but specifying `shape`

```python
with pymc3.Model() as model:
    rate = pymc3.Beta('rate', alpha=2, beta=2)
    input_var = pymc3.Data('input_var', [1, 2, 3])
    output_var = pymc3.Binomial('output_var', n=input_var, p=rate, shape=(3,))
pymc3.model_to_graphviz(model)
```

### Without observations and `shape` specification

```python
with pymc3.Model() as model:
    rate = pymc3.Beta('rate', alpha=2, beta=2)
    input_var = pymc3.Data('input_var', [1, 2, 3])
    output_var = pymc3.Binomial('output_var', n=input_var, p=rate)
pymc3.model_to_graphviz(model)
```
