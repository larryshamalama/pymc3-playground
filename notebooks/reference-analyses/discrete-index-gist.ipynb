```python
%matplotlib inline
import arviz as az
from matplotlib import pyplot as plt
import numpy as np
import pymc3 as pm
import seaborn as sns
```

```python
sns.set(color_codes=True)
SEED = 123456789

rng = np.random.default_rng(SEED)
```

```python
N = 200

MU = np.array([-2. , 2.])
W = np.array([0.75, 0.25])
```

```python
comp = rng.multinomial(1, W, size=N).argmax(axis=1)
x = rng.normal(MU[comp], 0.5)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.hist(x, bins=30, density=True);
ax.axvline(MU[0], c='k', ls='--');
ax.axvline(MU[1], c='k', ls='--');
```

```python
with pm.Model() as model:
    μ = pm.Normal("μ", 0, 5., shape=W.size,
                  transform=pm.transforms.ordered,
                  initval=np.arange(W.size) - W.size // 2)
    
    w = pm.Dirichlet("w", np.ones(W.size), shape=W.size)
    comp_ix = pm.Categorical("comp_ix", w, shape=N)
    
    obs = pm.Normal("obs", μ[comp_ix], 0.5, observed=x)
    trace = pm.sample(
        random_seed=SEED,
        return_inferencedata=True
    )
```

```python
_ = pm.plot_trace(trace)
```
